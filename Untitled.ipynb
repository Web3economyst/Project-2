```python
!pip install linearmodels


# ---- 1) Bibliotecas ----
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from linearmodels import PooledOLS
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from statsmodels.formula.api import ols
```

```python
# ---- 2) Carregar e filtrar dados ----
dados = pd.read_csv("sul_snis_municipio_agua_esgoto.csv")
```

```python
# ---- 4) HEATMAP DE MISSING VALUES ----
# ---- Verifica se as colunas necessárias existem ----
if all(col in dados.columns for col in ["sigla_uf", "ano"]):

    # ---- Seleciona apenas variáveis numéricas de interesse ----
    excluidas = ["id_municipio", "ano", "sigla_uf"]
    num_cols = [c for c in dados.select_dtypes(include=[np.number]).columns if c not in excluidas]

    # ---- Cálculo do percentual de missing por estado e ano ----
    missing_por_ano_uf = (
        dados.groupby(["sigla_uf", "ano"])[num_cols]
        .apply(lambda x: pd.Series({
            "total_missing": x.isna().sum().sum(),
            "total_campos": x.shape[0] * x.shape[1]
        }))
        .reset_index()
    )

    missing_por_ano_uf["perc_missing"] = np.where(
        missing_por_ano_uf["total_campos"] > 0,
        100 * missing_por_ano_uf["total_missing"] / missing_por_ano_uf["total_campos"],
        0
    )

    # ---- Heatmap de missing values ----
    if not missing_por_ano_uf.empty:
        pivot = missing_por_ano_uf.pivot(index="sigla_uf", columns="ano", values="perc_missing")

        plt.figure(figsize=(10, 6))
        sns.heatmap(
            pivot,
            cmap="plasma",
            annot=True,
            fmt=".1f",
            cbar_kws={"label": "% Missing"}
        )
        plt.title("Heatmap de Missing Values por Estado e Ano", fontsize=14)
        plt.xlabel("Ano")
        plt.ylabel("Estado")
        plt.tight_layout()
        plt.show()
    else:
        print("Nenhum dado disponível para gerar o heatmap.")
else:
    print("Variáveis 'sigla_uf' ou 'ano' não encontradas para análise de missing.")


# ---- Estatísticas descritivas (variáveis de interesse) ----
print("\nEstatísticas Descritivas (variáveis numéricas de interesse):\n")

num_vars = dados[[c for c in dados.select_dtypes(include=[np.number]).columns if c not in excluidas]]

if not num_vars.empty:
    desc_stats = num_vars.agg(["mean", "median", "std", "min", "max"]).T
    desc_stats = desc_stats.rename_axis("variavel").reset_index()
    print(desc_stats.head(20))  # mostra as 20 primeiras variáveis
else:
    print("Nenhuma variável numérica de interesse encontrada.\n")
```

```python
import plotly.express as px

# ---- 1️⃣ Selecionar variáveis de interesse ----
excluir = ["id_municipio", "sigla_uf", "ano"]
variaveis_analiticas = [c for c in dados.columns if c not in excluir]

# ---- 2️⃣ Calcular % de missing ----
missing_ratio = dados[variaveis_analiticas].isna().mean().sort_values(ascending=False) * 100

# ⚠️ Aqui: não removemos nenhuma coluna (mesmo 100% NA)
missing_df = pd.DataFrame({
    "Variável": missing_ratio.index,
    "% Missing": missing_ratio.values
})

# ---- 3️⃣ Destacar as variáveis com mais de 30% de missing ----
limite = 30
missing_df["Removida (>30%)"] = missing_df["% Missing"] > limite

# ---- 4️⃣ Gráfico interativo ----
fig = px.bar(
    missing_df,
    x="Variável",
    y="% Missing",
    color="Removida (>30%)",
    color_discrete_map={True: "crimson", False: "steelblue"},
    hover_data={"% Missing": ':.2f'},
    title=f"Percentual de Missing Values por Variável (limite = {limite}%)"
)

fig.update_layout(
    xaxis_title="Variáveis",
    yaxis_title="% de Missing",
    xaxis_tickangle=-45,
    template="plotly_white",
    height=600
)

fig.show()

# ---- 5️⃣ Excluir as variáveis acima do limite (>30%), exceto as de interesse ----
variaveis_interesse = ["populacao_atendida_agua", "populacao_atendida_esgoto"]

cols_to_drop = [
    col for col in missing_ratio[missing_ratio > limite].index
    if col not in variaveis_interesse
]

if cols_to_drop:
    dados_limpo = dados.drop(columns=cols_to_drop)
    print(f"{len(cols_to_drop)} variáveis removidas (> {limite}% missing):\n{cols_to_drop}")
else:
    dados_limpo = dados.copy()
    print("Nenhuma variável removida — todas têm menos de 30% de missing.")

# ---- 6️⃣ Exibir missing das variáveis de interesse ----
print("\n✅ Variáveis mantidas mesmo com missing alto:")
for var in variaveis_interesse:
    if var in dados.columns:
        miss = dados[var].isna().mean() * 100
        print(f"  • {var}: {miss:.1f}% de missing")
```

```python
# ---- 1️⃣ Selecionar apenas variáveis numéricas ----
dados_corr = dados_limpo.drop(columns=["sigla_uf", "ano", "id_municipio"], errors="ignore")
dados_num = dados_corr.select_dtypes(include=["float64", "int64"])

# ---- 2️⃣ Variáveis de interesse ----
vars_interesse = ["populacao_atendida_agua", "populacao_atendida_esgoto"]
vars_existentes = [v for v in vars_interesse if v in dados_num.columns]

if not vars_existentes:
    raise ValueError(f"Nenhuma das variáveis {vars_interesse} foi encontrada em dados_limpo.")

# ---- 3️⃣ Calcular correlação com cada variável de interesse ----
corr_df = pd.DataFrame()
for var in vars_existentes:
    corrs = dados_num.corrwith(dados_num[var]).sort_values(ascending=False).reset_index()
    corrs.columns = ["variavel", "correlacao"]
    corrs["referencia"] = var
    corr_df = pd.concat([corr_df, corrs], ignore_index=True)

# ---- 4️⃣ Remover autocorrelação ----
corr_df = corr_df[corr_df["variavel"] != corr_df["referencia"]]

# ---- 5️⃣ Ordenar para visualização ----
corr_df = corr_df.sort_values(by="correlacao", ascending=False)

# ---- 6️⃣ Gráfico interativo ----
fig = px.bar(
    corr_df,
    x="variavel",
    y="correlacao",
    color="referencia",
    color_discrete_map={
        "populacao_atendida_agua": "#1f77b4",
        "populacao_atendida_esgoto": "#d62728"
    },
    title="📊 Correlação com População Atendida (Água e Esgoto)",
    hover_data={"correlacao": ":.2f"},
)
fig.update_layout(
    xaxis_title="Variável",
    yaxis_title="Correlação de Pearson",
    xaxis_tickangle=-45,
    width=950,
    height=600,
    template="plotly_white",
)
fig.show()
```

```python
# 1️⃣ Estrutura de painel
dados_panel = dados_limpo.set_index(["id_municipio", "ano"])

# 2️⃣ Seleção das variáveis
target = "populacao_atendida_agua"
excluir = ["id_municipio", "ano", "sigla_uf", target]
X = dados_panel.drop(columns=[col for col in excluir if col in dados_panel.columns])
y = dados_panel[target]

# 3️⃣ Remove colunas constantes
const_cols = [c for c in X.columns if X[c].nunique() <= 1]
if const_cols:
    print("🔸 Removendo colunas constantes:", const_cols)
    X = X.drop(columns=const_cols)

# 4️⃣ Remove colunas altamente correlacionadas (r > 0.95)
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr = [column for column in upper.columns if any(upper[column] > 0.95)]
if high_corr:
    print("🔸 Removendo colunas altamente correlacionadas:", high_corr)
    X = X.drop(columns=high_corr)

# 5️⃣ Remove missing
dados_modelo = pd.concat([y, X], axis=1).dropna()
y = dados_modelo[target]
X = dados_modelo.drop(columns=[target])

# 6️⃣ Adiciona constante
X = sm.add_constant(X)

# 7️⃣ Estima modelo pooling
modelo_pooling = PooledOLS(y, X, check_rank=False)
resultado_pooling = modelo_pooling.fit()

# 8️⃣ Resultado
print(resultado_pooling.summary)
```
